```python
def decrypt_cipher(ciphertext,key):
    alphabet = string.ascii_lowercase
    plaintext = ciphertext.translate(str.maketrans(alphabet,key))
    return plaintext


def generate_di_freq_matrix(ciphertext, key):
    
    alphabet = string.ascii_lowercase
    plaintext = decrypt_cipher(ciphertext,key)

    freq_mat = np.zeros(shape=(26,26))

    for i in range(len(plaintext)-1):

        if (plaintext[i] in alphabet) and (plaintext[i+1] in alphabet):
            freq_mat[ord(plaintext[i+1])-97][ord(plaintext[i])-97]+=1
            
    freq_mat = freq_mat/freq_mat.sum()
    return freq_mat


def generate_trig_freq_matrix(ciphertext, key):
    
    alphabet = string.ascii_lowercase
    plaintext = decrypt_cipher(ciphertext,key)

    freq_mat = np.zeros(shape=(26,26,26))

    for i in range(len(plaintext)-2):

        if (plaintext[i] in alphabet) and (plaintext[i+1] in alphabet) and (plaintext[i+2] in alphabet):
            freq_mat[ord(plaintext[i])-97][ord(plaintext[i+1])-97][ord(plaintext[i+2])-97]+=1
            
    freq_mat = freq_mat/freq_mat.sum()
    return freq_mat

def generate_quad_freq_matrix(ciphertext, key):
    
    alphabet = string.ascii_lowercase
    plaintext = decrypt_cipher(ciphertext,key)

    freq_mat = np.zeros(shape=(26,26,26,26))

    for i in range(len(plaintext)-3):

        if (plaintext[i] in alphabet) and (plaintext[i+1] in alphabet) and (plaintext[i+2] in alphabet)and (plaintext[i+3] in alphabet):
            freq_mat[ord(plaintext[i])-97][ord(plaintext[i+1])-97][ord(plaintext[i+2])-97][ord(plaintext[i+3])-97]+=1
            
    freq_mat = freq_mat/freq_mat.sum()
    return freq_mat


def swap_chars(st,a,b):
    result = ""
    for i in st:
        if i == a:
            result += b
        elif i == b:
            result += a
        else:
            result +=i
    return result

def fswap_dig(freq_mat,a,b):
    freq_mat = freq_mat.copy()
    a_cord = ord(a)-97
    b_cord = ord(b)-97
    freq_mat[[a_cord,b_cord]]=freq_mat[[b_cord,a_cord]]
    freq_mat[:,[a_cord,b_cord]]=freq_mat[:,[b_cord,a_cord]]
    return freq_mat

def fswap_trig(freq_mat,a,b):
    freq_mat = freq_mat.copy()
    a_cord = ord(a)-97
    b_cord = ord(b)-97
    freq_mat[[a_cord,b_cord]]=freq_mat[[b_cord,a_cord]]
    freq_mat[:,[a_cord,b_cord]]=freq_mat[:,[b_cord,a_cord]]
    freq_mat[:,:,[a_cord,b_cord]]=freq_mat[:,:,[b_cord,a_cord]]
    return freq_mat

def get_score(freq_mat,ngram):
    return np.abs(freq_mat-ngram).sum()
```
