```python
import pandas as pd
```

```python
df1=pd.read_csv("5gram.tsv",sep="\t")
df2=pd.read_csv("4gram.tsv",sep="\t")
```

```python
df2=df2.dropna()
df1=df1.dropna()
```

```python
df2["*/1:4"]=df2["*/1:4"]/df2["*/1:4"].sum()
df2["*/-4:-1"]=df2["*/-4:-1"]/df2["*/-4:-1"].sum()
```

```python
df1["*/*"]=df1["*/*"]/df1["*/*"].sum()
```

```python
df1
```

```python
freq_df=dict()
for index,row in df2.iterrows():
    ky = row["4-gram"].lower()
    val1 = row["*/1:4"]
    val2 = row["*/-4:-1"]
    freq_df[ky+" "]=val1
    freq_df[" "+ky]=val2
```

```python
for index,row in df1.iterrows():
    ky = row["5-gram"].lower()
    val1 = row["*/*"]
    freq_df[ky]=val1
```

```python
freq_df["atio "]
```

```python
import pickle

with open("5gram_probs.pickle", "wb") as f:
    pickle.dump(freq_df,f)
```

```python
with open("5gram_probs.pickle", "rb") as f:
    freq_df2 = pickle.load(f)
```
